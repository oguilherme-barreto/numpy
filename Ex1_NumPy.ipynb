# Tratando Erros

```python
plt.plot(datas, Kaliningrad)
```

```python
indice = np.where(np.isnan(Kaliningrad))

for i in indice[0]:
  Kaliningrad[i] = (Kaliningrad[i-1] + Kaliningrad[i+1]) / 2

Kaliningrad
```

```python
plt.plot(datas, Kaliningrad)
```

# Regressão Linear

```python
X = datas
Y = Moscow
n = Moscow.size

a = (n*sum(X*Y) - (sum(X)*sum(Y))) / ((n*sum(X**2)) - (sum(X))**2)

b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X + b

np.linalg.norm(Moscow - y)
```
